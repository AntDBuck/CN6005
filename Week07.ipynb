**Task: Accident Prediction**

P(C | X) = P(X | C) * P(C)  

Evidence, P(X), can be ignored as it is the same for all values

P(YES) = 5/10 = 1/2

P(NO) = 5/10 = 1/2

---

P(Rain | Yes) = 1/5

P(Good | Yes) = 1/5

P(Normal | Yes) = 1/5

P(No Engine Problem | Yes) = 2/5

---

P(Rain | NO) = 2/5

P(Good | NO) = 3/5

P(Normal | NO) = 2/5

P(No Engine Probelm | NO) = 4/5

---

P(X | YES) = 1/5 * 1/5 * 1/5 * 2/5 = 0.0032

P(X | NO) = 2/5 * 3/5 * 2/5 * 4/5 = 0.0192

---

P(YES | X = (Rain, Good, Normal, No Engine Problem))

= 0.0032 * 0.5 = 0.0016

P(NO | X = (Rain, Good, Normal, No Engine Problem))

= 0.0192 * 0.5 = 0.0096

0.0096 > 0.0016

---

Given the conditons of Rain, Good, Normal, No Engine Problem, the incident is classified as NO accident.

```python
# Create dataframe that houese data
df = pd.DataFrame(data)
df.head()
```

```python
# Convert label data into numerical values
df['label'] = df['label'].map({'spam': 1, 'not spam': 0})
df.head()
```

```python
# Split data into training and testing sets
X = df['text']
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Count words
vectoriser = CountVectorizer()
X_train_vectors = vectoriser.fit_transform(X_train)
X_test_vectors = vectoriser.transform(X_test)
```

```python
# Create Naive Bayes Classifier model
model = MultinomialNB()
model.fit(X_train_vectors, y_train)
```

```python
# Make predictions and print evaultion metrics
y_pred = model.predict(X_test_vectors)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy score: {accuracy * 100:.0f}%')
print(f'Precision score: {precision * 100:.0f}%')
print(f'Recall score: {recall * 100:.0f}%')
print(f'F1 score: {f1 * 100:.0f}%')
```

```python
# Create new entries
new_msgs = [
    "Your free prize is waiting",
    "Are we still meeting tomorrow?",
    "Click now to claim your reward"
    ]

new_mgs_vectors = vectoriser.transform(new_msgs)
```

```python
# Predict label of new messages
pred_msgs = model.predict(new_mgs_vectors)

print(f'Message 1 was predicted as {'spam' if pred_msgs[0] == 1 else 'not spam'}')
print(f'Message 2 was predicted as {'spam' if pred_msgs[1] == 1 else 'not spam'}')
print(f'Message 3 was predicted as {'spam' if pred_msgs[2] == 1 else 'not spam'}')
```
