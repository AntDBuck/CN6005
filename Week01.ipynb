**List Example**

```python
# Create a list and print an f-string with list length.
a = [1, 2, 4, 5, 6, 14, 25]
print(f'Total members in the list is {len(a)}.')
print(a[2])
# Change element at index 5, then print it.
a[5] = 17
print(a[5])
# Print smallest and largest values in list.
print(min(a), max(a))
# Add new element to end of list, then print f-string with list length.
a.append(1500)
print(f'Total members in the list is {len(a)}.')
print(a[7])
```

**Local and Global Variables**

```python
gVariable = 10
# Define new function with one parameter.
def listSum(numList):
  # Set global variable and its value, then print f-string.
  global gVariable
  gVariable = 20
  print(f'Value of the gVariable is {gVariable}')
  theSum = 0
  # Loop through all members in list and add each to theSum.
  for i in numList:
    theSum = theSum + i
  count = len(numList)
  print(f'Count of the list is {count}')
  return theSum, count
```

```python
# Testing scope of variable.
print(f'Value of gVariable before function call: {gVariable}')
myList = [1, 3, 6, 9, 10, 15, 14, 8]
# Notice the position of name declaration with respect to the returned values.
sumValue, countValue = listSum(myList)
print(f'Sum of the list is {sumValue}')
print(f'Count of the list is {countValue}')
print(f'Value of gVariable after function call: {gVariable}')
```

```python
a = []
even = []
odd = []
for i in range(1, 101, 1):
  # If item is divisible by 2 in range then add to even list.
  if(i % 2 == 0):
    even.append(i)
    # Else item must be odd, add it to odd list.
  else:
    odd.append(i)
  a.append(i)
print(f'Total members of the list is {len(a)}.')
print(f'Total members of the even list is {len(even)}.')
print(f'Total members of the odd list is {len(odd)}.')
```

**Use Of Pandas**

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
# Create data frame using loaded iris data.
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
print(df.shape)
print(df.head())
```

```python
# Add new column to data frame.
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
print(df.head())
```

**Using Matplotlib**

```python
import matplotlib.pyplot as plt
# Create histogram from data frame.
plt.hist(df['sepal length (cm)'], bins = 20, color = 'pink', edgecolor = 'black')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

**Using Seaborn**

```python
import seaborn as sns
# Create boxplot from data frame.
sns.boxplot(x = 'species', y = 'petal width (cm)', data = df, color = 'green')
plt.title('Petal Width by Species')
plt.show()
```

```python
# Create scatterplot from data frame.
sns.scatterplot(x = 'petal length (cm)', y = 'petal width (cm)', hue = 'species', data = df)
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
# Do not include species.
corr = df.iloc[:, :-1].corr()
# Create heatmap from data frame.
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
grouped_df = df.groupby('species').mean()
print(grouped_df)
```
